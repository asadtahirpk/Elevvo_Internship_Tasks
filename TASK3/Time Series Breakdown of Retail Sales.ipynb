```python
import pandas as pd 
```

# Load Data_Set

```python
features=pd.read_csv("features.csv")
stores=pd.read_csv("stores.csv")
train=pd.read_csv("train.csv")
```

# Merge  Data

```python
train_stores=pd.merge(train,stores,on="Store",how="left")
```

```python
train_stores.head()
```

```python
train_stores.tail()
```

```python
master_data = pd.merge(train_stores, features, on=["Store", "Date"], how="left")
```

```python
master_data.head()
```

```python
master_data.tail()
```

```python
master_data.info()
```

# Drop Uncessary Columns 

```python
master_data = master_data.drop(['Type', 'Size', 'Fuel_Price', 'MarkDown1', 'MarkDown2',  'MarkDown3', 'MarkDown4', 'MarkDown5', 'CPI', 'Unemployment', 'IsHoliday_y'], axis=1)
```

```python
master_data.head()
```

```python
master_data.tail(30)
```

```python
master_data.info()
```

# Clean Data 

```python
master_data=master_data.rename(columns={"IsHoliday_x":"IsHoliday"})
```

```python
master_data["Date"]=pd.to_datetime(master_data["Date"])
```

```python
master_data.info( )
```

```python

print("Data set Shape: ",master_data.shape)
```

```python
print("\nSummary Statistics:")
master_data.describe()
```

```python
print("\nWeekly_Sales median")
master_data["Weekly_Sales"].median()
```

# Monthly Sales Trend Analysis 

```python
master_data.head()
```

```python
master_data["Month_Year"]=master_data["Date"].dt.to_period("M")
```

```python
master_data.head()
```

```python
monthly_sales=master_data.groupby("Month_Year")["Weekly_Sales"].sum().reset_index()
monthly_sales.rename(columns={"Weekly_Sales":"Monthly_Sales"},inplace=True)
monthly_sales.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.lineplot(
    data=monthly_sales,
    x=monthly_sales['Month_Year'].astype(str),
    y='Monthly_Sales',
    marker="o",
    label="Monthly Sales"
)

plt.xticks(rotation=45)
plt.title("Monthly Sales Trend of Welmart")
plt.xlabel("Month")
plt.ylabel("Monthly Sales")
plt.show()

```

# Moving Average 

```python
monthly_sales["3M_MA"]=monthly_sales["Monthly_Sales"].rolling(window=3).mean()
```

```python
monthly_sales.head()
```

```python
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.lineplot(
    data=monthly_sales,
    x=monthly_sales['Month_Year'].astype(str),
    y='Monthly_Sales',
    marker="o",
    color="blue",
    label="Monthly Sales"
)
sns.lineplot(
    data=monthly_sales,
    x=monthly_sales['Month_Year'].astype(str),
    y='3M_MA',
    marker="o",
    color="red",
    label="3_Month_MA"
)
plt.xticks(rotation=45)
plt.title("Monthly Sales with 3-Month Moving Average")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Seasonality Analysis

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
monthly_sales.index=pd.to_datetime(monthly_sales["Month_Year"].astype(str))
```

```python
monthly_sales.head()
```

```python

result = seasonal_decompose(monthly_sales['Monthly_Sales'], model='additive', period=12)
# Plot trend, seasonal, residual
result.plot()
plt.show()
```

# Region-wise Breakdown

```python
product_sales = master_data.groupby(['Dept', 'Month_Year'])['Weekly_Sales'].sum().reset_index()
product_sales.rename(columns={"Weekly_Sales":"Monthly_Sales"},inplace=True)
product_sales.head()
```

# Total Sales per Dept in a month

```python
total_sale_per_Dept=product_sales.groupby("Dept")["Monthly_Sales"].sum().reset_index()
total_sale_per_Dept.head()
```

# Top Five Dept by Total Sales

```python
top_Dept=total_sale_per_Dept.nlargest(5,"Monthly_Sales")
top_Dept.head()
```

```python
top_Dept.rename(columns={"Monthly_Sales":"Total_Sales"},inplace=True)
```

```python
top_Dept.head()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Dept",y="Total_Sales",data=top_Dept)
plt.xlabel("Top 5 Department")
plt.ylabel("Total Sales Per Department")
plt.title("Top 5 Department By Total Sales")
plt.show()
```

```python
product_sales.head()
```

# Top 5 stores By Total Sales 

```python
top_Dept_list = (product_sales.groupby('Dept')['Monthly_Sales']
                            .sum()
                            .nlargest(5)
                            .index
                            .tolist())
top_Dept_list
```

# Filter monthly sales of  Top 5 Department 

```python
filtered = product_sales[product_sales['Dept'].isin(top_Dept_list)].copy()
```

```python
filtered.head()
```

```python
filtered["Dept"].unique()
```

```python
filtered["Month_Year"]=filtered["Month_Year"].dt.to_timestamp()
```

# Sales Trends of Top 5 Department

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered, x='Month_Year', y='Monthly_Sales', hue='Dept', marker='o')
plt.legend(title="Department", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Product-wise Monthly Sales Trend Of Top 5 Department")
plt.xlabel("Months")
plt.show()
```

```python
master_data.head()
```

# Region-wise Breakdown

```python
region_sales=master_data.groupby(["Store","Month_Year"])["Weekly_Sales"].sum().reset_index()
region_sales.rename(columns={"Weekly_Sales":"Monthly_Sales"},inplace=True)
region_sales.head()
```

# Top 5 stores By Total Sales 

```python
top_Store_list = (region_sales.groupby('Store')['Monthly_Sales']
                            .sum()
                            .nlargest(5)
                            .index
                            .tolist())
top_Store_list
```

# Filter monthly sales of  Top 5 Stores 

```python
filtered2 = region_sales[region_sales['Store'].isin(top_Store_list)].copy()
filtered2.head()
```

```python
filtered2["Month_Year"]=filtered2["Month_Year"].dt.to_timestamp()
```

# Sales Trends of Top 5 store

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered2, x='Month_Year', y='Monthly_Sales', hue='Store', marker='o')
plt.legend(title="Stores", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("Months")
plt.title("Region-wise Monthly Sales Trend Of Top 5 Stores")
plt.show()
```
